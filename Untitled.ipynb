```python
import numpy as np
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
tf.__version__
```

```python
#image processing
```

```python
# ImageDataGenerator: This is a class from Keras used for generating batches of tensor image data with real-time data augmentation.
#training_set': This is the directory where the training images are located
```

```python
    train_datagen = ImageDataGenerator(
            rescale=1./255,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True)
    training_set = train_datagen.flow_from_directory(
            'training_set',
            target_size=(64, 64),
            batch_size=32,
            class_mode='categorical')
```

```python
# test_datagon:This is the directory where the test images are located.
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        'test_set',
        target_size=(64, 64),
        batch_size=32,
        class_mode='categorical')
```

```python
cnn = tf.keras.models.Sequential() #data flow will be sequential with the keras model 
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=64 , kernel_size=3 , activation='relu' , input_shape=[64,64,3])) 
# This adds a 2D convolutional layer to the model.
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2,strides=2))
#adds a max-pooling layer to our convolutional neural network (CNN) model. 
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=64 , kernel_size=3 , activation='relu' ))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2 , strides=2))
```

```python
cnn.add(tf.keras.layers.Dropout(0.5)) # dropout:  commonly used in neural networks to prevent overfitting
```

```python
cnn.add(tf.keras.layers.Flatten()) # covert the 3d model to the 1d moedel
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu')) #negative value willl be =0
```

```python
cnn.add(tf.keras.layers.Dense(units=5 , activation='softmax'))
```

```python
cnn.compile(optimizer = 'rmsprop' , loss = 'categorical_crossentropy' , metrics = ['accuracy'])
```

```python
cnn.fit(x = training_set , validation_data = test_set , epochs = 30)
```

```python
from keras.preprocessing import image
test_image =tf.keras.utils.load_img('prediction/r.jpg',target_size=(64,64))
test_image =tf.keras.utils.img_to_array(test_image)
test_image = np.expand_dims(test_image,axis=0)
result = cnn.predict(test_image)
training_set.class_indices
```

```python
categories = ["daisy", "dandelion", "rose", "sunflower" , "tulip"] 
```

```python
import matplotlib.pyplot as plt
predicted_class = categories[np.argmax(result)]
test_image =tf.keras.utils.load_img('prediction/r.jpg',target_size=(200,200))
plt.imshow(test_image)
plt.title(predicted_class)  
plt.show()
#test_image = tf.keras.utils.img_to_array(test_image)
#test_image = np.expand_dims(test_image,axis=0)
```
